```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# model = YOLO("best.pt")
model = YOLO("/media/akashs/DATA/MyGithub/AI_ML_DS_projects/Face_Mask_Detection/archive/runs/detect/train2/weights/best.pt")
img = cv2.imread("tes_image.jpg")
results = model(img)
# results[0].show()  # Opens window with detection
```

```python

# Get the result image (NumPy array with bounding boxes drawn)
result_img = results[0].plot()

# Convert BGR (OpenCV default) to RGB for matplotlib
result_img_rgb = cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB)

# Display in notebook
plt.figure(figsize=(20, 8))
plt.imshow(result_img_rgb)
plt.axis("off")
plt.title("YOLO Detection")
plt.show()
```
